```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd
from scipy.stats import f_oneway
from statsmodels.sandbox.stats.multicomp import multipletests
import numpy as np
```

```python
counts=pd.read_table('../Data/Done_deal.txt',index_col=0)
hpa=pd.read_table('../Data/proteinatlas.tsv')
counts=counts.loc[hpa.Ensembl]
counts=counts.dropna()
# counts.loc['conds']=[i[0] for i in counts.columns]
gene_length=pd.read_table('../Data/gene_length.txt',index_col=0)
gene_length=gene_length.loc[counts.index]
counts=np.log2(rpkm(counts,gene_length)+1)
```

```python
def rpkm(counts, lengths):
    N = counts.sum()
    L = lengths.loc[counts.index]['Length'].map(float)
    C = counts
    normed = 1e9*((C.T/L).T/N)
    return(normed)
def tpm(counts, length):
    rate=(counts.T/length.loc[counts.index]['Length'].map(float)).T
    denom=rate.sum()
    normed=rate*1e6/denom
    return normed
```

```python
anova=f_oneway(counts.T[[i[0]=='N' for i in counts.columns]],counts.T[[i[0]=='S' for i in counts.columns]],counts.T[[i[0]=='H' for i in counts.columns]])#,counts.T[[i[0]=='C' for i in counts.columns]])
anova=pd.DataFrame(zip(list(counts.index),anova[1]),columns=['gene','pvalue']).dropna()
anova['padj']=multipletests(anova['pvalue'],method='fdr_bh')[1]
anova=hpa[['Gene','Ensembl']].merge(anova,how='left',right_on='gene',left_on='Ensembl').sort_values('padj')
anova[anova['padj']<0.05].to_csv('haha.csv')
```
